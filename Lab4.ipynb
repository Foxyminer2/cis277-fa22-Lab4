## NFL Offense Data

1. Import the NFL week 3, 4, and 5 data and add a week column to dataset and set the value to 3, 4, or 5 for all rows from that week.

```python
pd.read_csv('nfloffenseweek3.csv').assign(Week=3)
pd.read_csv('nfloffenseweek4.csv').assign(Week=4)
pd.read_csv('nfloffenseweek5.csv').assign(Week=5)
# TODO
```

2. Combine the weekly dataframes by adding the dataframes together row-wise.

```python
offenses = sorted(glob('nfloffenseweek*.csv'))
df = pd.concat((pd.read_csv(file) for file in offenses), ignore_index = True)
df.columns
```

3. Make up 1 interesting question about this dataset. Answer it.

Which teams have greater than or equal to 1200 total yards, scored over 80 points so far, and have less than 20 penalties?

```python
pd.DataFrame(df[(df["total_yards"] >= 1200.0) & (df["points_scored"] >= 80) & (df['penalties'] < 20)].team)
```

## World Population Data

4. Join the height and weight data to the iq data. 
5. Join population density to quality of life data.
6. Join life expectancy to quality of life data.
7. Join iq, life expectancy, quality of life and population density into 1 dataframe.

```python
health = pd.read_csv('height_weight_data.csv')
iq = pd.read_csv('iq.csv')
population = pd.read_csv('population_density.csv')
quality_of_life = pd.read_csv('quality_of_life.csv')
life_expectancy = pd.read_csv('life_expectancy.csv')
```

4. Join the height and weight data to the iq data. 

```python
health.merge(iq,left_on='country', right_on='country', how='right')
```

5. Join population density to quality of life data.

```python
population.merge(quality_of_life,left_on='country', right_on='country', how='right')
```

6. Join life expectancy to quality of life data.

```python
life_expectancy.merge(quality_of_life,left_on='country', right_on='country', how='right')
```

7. Join iq, life expectancy, quality of life and population density into 1 dataframe.

```python
full_dataset = ((iq.merge(life_expectancy,left_on='country', right_on='country', how='right')).merge(quality_of_life,left_on='country', right_on='country', how='left')).merge(population,left_on='country', right_on='country', how='left')
full_dataset
```

8. Make up 1 interesting question about one of these datasets and answer it.
What are the top country's based on IQ and both male and female life expectancy?

```python
pd.DataFrame(full_dataset[(full_dataset["iq"] >= 100.0) & (full_dataset["male_life_expectancy"] >= 80.0) & (full_dataset['female_life_expectancy'] >= 80.0)].country)
```

9. Make up 1 interesting question about one of these datasets and answer it.
What the average birth rate and death rate of each country?

```python
pd.DataFrame(full_dataset.groupby('country').agg({'birth_rate':['mean'],'death_rate': ['mean']}))
```

10. Make up 1 interesting question about one of these datasets and answer it.
Which countries are the most popular and most safe over 70?

```python
pd.DataFrame(full_dataset[(full_dataset["popularity"] >= 70.0) & (full_dataset["safety"] >= 70.0)].country)
```
